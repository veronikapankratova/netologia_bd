```python
import re
import traceback
```

```python
# А, В, Е, К, М, Н, О, Р, С, Т, У и Х.

string = 'С234АА23'
try:
    pattern =r'([АВЕКМНОРСТУХ]{1}\d{3}[АВЕКМНОРСТУХ]{2})(\d{2,3})'
    result = re.search(pattern, string)
# for items
    print ('номер транспортного средства', result.group(1), 'регион', result.group(2))
except Exception:
    print(traceback.print_exc())
```
