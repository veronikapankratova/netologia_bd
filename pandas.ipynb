```python
# Задача №1

# https://grouplens.org/datasets/movielens/

ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
movies = pd.read_csv('ml-latest-small/movies.csv')
movies.head()
```

```python
joined = ratings.merge(movies, on='movieId', how='left')
joined.head()
```

```python
joined['rating'].value_counts().head(10)
```

```python
movie_stats = joined.filter(items = ['title', 'rating'])
movie_stats.head()
```

```python
# фильтр на несколько условий сразу
# | - условие ИЛИ
# & AND
# (() | () | ()) & ()
filtered_movies = movie_stats[ (movie_stats['rating']== 5)]
filtered_movies.head(10)
```

```python
# filtered_movies.sort_values('rating', ascending=False).head(1)
filtered_movies['title'].value_counts().head(1)
```

```python
# Задача №2
# По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония)
# категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.



data = pd.read_csv('power.csv')
data.head()
```

```python
# фильтр на несколько условий сразу
# | - условие ИЛИ
# & AND
# (() | () | ()) & ()
filtered_countries = data[ ((data['country']=='Latvia') | (data['country']=='Lithuania')| (data['country']=='Estonia'))
                          &((data['year'] > 2005) | (data['year'] < 2010))
                          &(data['quantity'] > 0)
                          &((data['category'] == 4)| (data['category'] == 12)| (data['category'] == 21)) ]

filtered_countries.head(500)
```
